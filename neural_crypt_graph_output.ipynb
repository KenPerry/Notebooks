```python
import sys
print( sys.version )
# Need added "encoding" arg. to read file generate by Python 2 in Python 3
p = pickle.load( open('bob_eve_pickle', 'rb'), encoding='latin1' )
bob_errors, eve_errors = p[0], p[1]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("darkgrid")

plt.plot(bob_errors)
plt.plot(eve_errors)
plt.legend(['bob', 'eve'])
plt.xlabel('Epoch')
plt.ylabel('Lowest Decryption error achieved')
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(bob_errors, 'b')
ax1.set_ylabel('Bob errors')

ax2 = ax1.twinx()
ax2.plot(eve_errors, 'r')
ax2.set_ylabel('Eve errors')
```

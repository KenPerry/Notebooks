```python
# Load the titanic data
# the dataset is now a Pandas DataFrame
titanicData = pandas.read_csv("./Titanic.csv")
```

```python
# Start by adding extra columns if required

# Add Variable for total family members
titanicData['NumFamilyMembers'] = titanicData['NumSiblingsSpouse'] + titanicData['NumParentsChildren']

# Add bins of age
age_bins = numpy.arange(0,titanicData['Age'].max(), 10);

age_labels = [];
for i in range(len(age_bins)):
    if i < len(age_bins)-1:
        age_labels.append((age_bins[i]+age_bins[i+1])/2)

titanicData['AgeBins'] = pandas.cut(titanicData['Age'], age_bins, labels=age_labels);
#titanicData['AgeBins'] = titanicData['AgeBins'].replace(numpy.nan, -1.0)

# Also add AgeGroups
age_groups = [0, 18, 50, 100]
age_group_names = [ 0, 1, 2 ]
titanicData['AgeGroups'] = pandas.cut(titanicData['Age'], age_groups, labels=age_group_names);
#titanicData['AgeGroups'] = titanicData['AgeGroups'].replace(numpy.nan, -1.0)
```

```python
# Just retain the relevant variables
dependent    = 'Survived';
independents = [ 'GenderCode', 'PassengerClass',
                 'NumFamilyMembers', 'NumSiblingsSpouse', 'NumParentsChildren',
                 'AgeBins', 'AgeGroups', 'Age'
               ]
allVars = [ dependent ] + independents;

titanicData = titanicData[allVars]

# See Columns in the DataSet
HTML("<b>Dependent Variables:<b>")
print(dependent)

HTML("<b>Independent Variables:<b>")
print(independents)
```

```python
# See sample data in admitData DataFrame
HTML("<b>Sample Data:<b>")
titanicData.head()

# Show number of rows in admitData
HTML("<b>Row Count:<b>")
len(titanicData)

# Describe the data
# Show non-null count, mean, std dev, min, max, etc
HTML("<b>Basic Data Description:<b>")
titanicData.describe().round(2)

# See distribution of each variable
HTML("<b>Distribution of Input Data:<b>")
count = 0;
for varName in allVars:
    uniqueVals = titanicData[varName].unique()
    
    if len(uniqueVals) <= 10:
        valsToPlot = titanicData[varName].value_counts(normalize=True,dropna=False)*100
    else:
        valsToPlot = pandas.cut(titanicData[varName], 10).value_counts(normalize=True,dropna=False).round(2) * 100

    if count == 0:
        fig, axes = matplot.subplots(nrows=1, ncols=3, figsize=(10,3))
           
    plot = valsToPlot.sort_index().plot(kind='bar', ax = axes[count], title=varName)
   
    if count < 2:
        count = count + 1;
    else:
        count = 0

matplot.show()         
HTML("<b>Things To Note:</b>"
     +nl+"* There are a large number of missing ages"
     +nl+"* There are not many passengers having more than 3 siblings or spouses or more than 3 family members"
    )
```

```python
HTML("<b>From the distributions above, we notice that there are a few bins with very little data<br>We should combine them into single bucket (ex: NumParentsChildren>=3):<b>")

# NumFamilyMembers >= 3
# NumSiblingsSpouse >= 2
# NumParentsChildren >= 3

# Truncate col values
count = 0;
for varName in ['NumFamilyMembers', 'NumSiblingsSpouse', 'NumParentsChildren']:
    
    titanicData[varName][titanicData[varName] >= 3] = 3
    valsToPlot = titanicData[varName].value_counts(normalize=True,dropna=False)*100
    
    if count == 0:
        fig, axes = matplot.subplots(nrows=1, ncols=3, figsize=(10,3))
      
    plot = valsToPlot.sort_index().plot(kind='bar', ax = axes[count], title=varName)

    if count < 2:
        count = count + 1;
    else:
        count = 0
```

```python
for varName in independents:
    if varName == 'Age':
      continue
    pctsToPlot = titanicData.replace(numpy.nan, -1).groupby(varName)[dependent].sum()/titanicData.replace(numpy.nan, -1).groupby([varName]).size();
    uniqueVals = titanicData[varName].unique();
    fig, axes = matplot.subplots(nrows=1, ncols=2, figsize=(10,3));
    plot1 = seaborn.violinplot(x=varName, y=dependent, ax = axes[0], data=titanicData);
    plot2 = pctsToPlot.plot(kind='bar', ax = axes[1], ylim=(0,1));
    title = fig.suptitle('Distribution of % '+dependent+' by '+varName, fontweight="bold", fontsize=12);
    matplot.show()

seaborn.FacetGrid(titanicData, hue=dependent, dropna=True, aspect=2, xlim=(0, 80)).map(seaborn.kdeplot,'Age',shade= True).add_legend()
matplot.title('Kernel density estimate of Survived vs Agr\nSeems like below the age of 10 there was a slightly higher probability of surviving\nOthers was a coin flip', fontweight="bold");
```

```python
colors = [ "red", "green", "blue", "yellow"];

interestingGroupsToPlot = [ ["GenderCode", "PassengerClass" ],
                            ["GenderCode", "AgeGroups" ],
                            ["GenderCode", "AgeBins" ],
                            ["GenderCode", "NumFamilyMembers" ] ,
                            ["GenderCode", "NumSiblingsSpouse" ],
                            ["GenderCode", "NumParentsChildren" ] ,
                            ["PassengerClass", "NumFamilyMembers" ],
                            ["PassengerClass", "NumSiblingsSpouse" ],
                            ["PassengerClass", "NumParentsChildren" ] 
                          ];

for group in interestingGroupsToPlot:
    pivotCond = pandas.pivot_table(titanicData.replace(numpy.nan, -1), values=dependent, 
                                   index=[group[0]], columns=[group[1]], aggfunc=numpy.mean
                                  ).dropna(axis=0,how='all').dropna(axis=1,how='all');
    pivotUncond = pandas.pivot_table(titanicData.replace(numpy.nan, -1), values=dependent, 
                                     index=[group[0]], columns=[group[1]], 
                                     aggfunc=numpy.sum
                                    ).dropna(axis=0,how='all').dropna(axis=1,how='all');

    fig, axes = matplot.subplots(nrows=1, ncols=4, figsize=(15,4));
    title = fig.suptitle('Distribution of % '+dependent+' by '+group[0]+" and "+group[1], fontweight="bold", fontsize=12);
    plot1 = seaborn.heatmap(pivotCond, cmap="YlGnBu", ax = axes[1]);
    plot3 = seaborn.heatmap(pivotUncond, cmap="YlGnBu", ax = axes[3]);
    ax1 = axes[1].invert_yaxis();
    ax3 = axes[3].invert_yaxis();
    #ax1 = axes[1].set_xticklabels(axes[1].xaxis.get_majorticklabels(), rotation=45)
    #ax3 = axes[3].set_xticklabels(axes[1].xaxis.get_majorticklabels(), rotation=45)
    var1Vals = pivotCond.index.get_level_values(0).unique()
    thisCount = 0;
    for var1Val in var1Vals:
        splitCond=pivotCond.xs(var1Val);
        splitUncond=pivotUncond.xs(var1Val);
        plot0 = splitCond.plot(kind="line",color=colors[thisCount], ax=axes[0], marker="o", ylim=(0,1)).legend();
        plot2 = splitUncond.plot(kind="line",color=colors[thisCount], ax=axes[2], marker="o").legend();
        #ax0 = axes[0].set_xticklabels(axes[1].xaxis.get_majorticklabels(), rotation=90)
        #ax2 = axes[2].set_xticklabels(axes[1].xaxis.get_majorticklabels(), rotation=90)
        thisCount = thisCount+1;
    matplot.show()
```

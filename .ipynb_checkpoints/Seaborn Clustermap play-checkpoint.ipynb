# We will use sk-learn to try to replicate sns.clustermap
## Looks like clustermap uses
###   Z = linkage(X, method='single', metric='euclidean')
## It is easier to see this if we reduce the number of points

```python
# needed imports
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
# generate the linkage matrix
Z = linkage(X, method='single', metric='euclidean')
```

## Show the step by step clustering
### These should correspond to the vertical groups, whose indices are on right side of clustermap
## The format of the linkage array:
### format [idx1, idx2, dist, sample_count]
### so this says we will cluster samples with index idx1 and idx2
### If either idx > numSamples
### then the idx represents a previous cluster, located at line Z[idx - len(X)]

Z

## Extract the clusters

```python
from scipy.cluster.hierarchy import fcluster
k=3

clusters = fcluster(Z, k, criterion='maxclust')
for clusterNum in range(1,k+1):
    # n.b., np.where returns a tuple, whose first element if the array of indices
    memIdxs = np.where(clusters == clusterNum)[0]
    
    print("Cluster {c}: count {cnt}, {members}.\n".format(c=clusterNum, cnt=len(memIdxs), members=memIdxs))
    
```

```python
import matplotlib.pyplot as plt
plt.ioff()
```

```python
# calculate full dendrogram
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dd = dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
dd.keys()
dd['leaves']

# plt.show()
```

```python
import pandas as pd
import seaborn as sns
sns.set()

# Load the brain networks example dataset
df = sns.load_dataset("brain_networks", header=[0, 1, 2], index_col=0)
```

```python
df.head()
len(df)
df.shape
df.index
df.describe()
```

```python
# Select a subset of the networks
used_networks = [1, 5, 6, 7, 8, 12, 13, 17]
used_columns = (df.columns.get_level_values("network")
                          .astype(int)
                          .isin(used_networks))
df = df.loc[:, used_columns]

# Create a categorical palette to identify the networks
network_pal = sns.husl_palette(8, s=.45)
network_lut = dict(zip(map(str, used_networks), network_pal))

# Convert the palette to vectors that will be drawn on the side of the matrix
networks = df.columns.get_level_values("network")
network_colors = pd.Series(networks, index=df.columns).map(network_lut)
```

```python
plt.ion()
```

```python
# Draw the full plot
sns.clustermap(df.corr(), center=0,  # cmap="vlag",
               row_colors=network_colors, col_colors=network_colors,
               linewidths=.75, figsize=(13, 13))
```
